## <font color='blue'>**I. Descripción del problema**</font>

La **tasa de desempleo** es sin duda uno de los índices que más atención e interés concentra pues afecta de manera directa a toda la nación de manera transversal. Cada mes se ve cómo en los medios de comunicación, una de las principales noticias es efectivamente mostrar este índice actualizado.

 Una **elevada tasa de desocupación conlleva una serie de efectos**, tanto en la economía como en la sociedad misma. Por una parte el tener menos puestos de trabajo probablemente implique una menor producción y crecimiento económico. A su vez, las familias al no contar con un salario comenzarán a disminuir su gasto y consumo por lo que el flujo de dinero se ralentizará, y en consecuencia el Estado deberá aumentar ante este escenario su gasto social para asistir a los afectados. Todo lo anterior evidentemente acarrea una serie de dificultados psicológicas en las personas debido al alto estrés que implica las dificultades financieras en los individuos y las familias.

 El desempleo sin duda es un **fenómeno multifactorial**, que está ligado tanto a eventos periódicos (como rendimiento de la economía y expectativas) así como a eventos coyunturales (desastres naturales o conflictos sociales,  por ejemplo).



En este proyecto nos centraremos en la **tasa de desocupación en Chile** y trataremos de explicarla a través de distintos factores (periódicos o recurrentes) que creemos deberían estar fuertemente ligados a este indicador.

## <font color='blue'>**II. La Pregunta de Investigación e Hipótesis**</font>

Como se mencionó en el punto anterior, el desempleo es un fenómeno multicausal que es influenciado por diversos factores. En particular, nos interesa saber cómo las variables asociadas a **actividad económica** y la **TPM**(tasa de política monetaria) impactan en este indicador, pues la intuición lleva a pensar que estas variables son de las que más inciden en el nivel de desocupación.


De esta manera, nuestar hipótesis inicial es que el desempleo en Chile puede ser en gran medida explicado mediante un modelo de regresión lineal múltiple del tipo:






$ Desempleo_{m} = f(actividad económica_{m} , TPM_{m})  + {\epsilon}_{m} $

Es decir, se tratará de explicar la tasa de desempleo en el mes m como una función de las variables relacionadas a actividad económica y TPM (ambas en mes m) ademas de una parte no sistemática que considera la aleatoriedad del fenómeno.

En esta caso **la hipótesis inicial es el nivel de actividad económica del país y la TPM influyen fuertemente en el nivel de desempleo.**

Lo primero, pues es esperable que el país al producir más, requiera más puestos de trabajo para mantener dicho nivel productivo. Lo segundo, debido a que un aumento en la TPM provoca aumento en los costos de endeudamiento, lo que puede derivar en una desaceleración en gasto e inversión, y con ello e requerir menos puestos de trabajo.

Hemos logrado especificar el modelo al plantear:
- El error aleatorio como parte inherente del fenómeno
- Definido la forma funcional de tipo lineal

En cuanto a la métrica de las variables, se verá en detalle en el siguiente punto "Descripción de la data".

A su vez también se propondrá enriquecer el modelo anteriormente planteado mediante variables relacionadas a:

- Expectativas económicas
- Volumen de avisos de empleo

Que también se cree puedan estar en alta sintonía con el nivel de desocupación de un país.

### <font color='blue'>1. Variables utilizadas</font>

Para resolver la pregunta de investigación es crucial con la data adecuada. En este caso, tanto la variable dependiente como las variables independientes y potenciales regresores, se extraerán del repositorio de datos del Banco Central:

Se extraerán los indicadores con periodicidad mensual, desde Enero 2018 a Agosto 2023 (68 observaciones).

Las variables con las que se trabajará son:

####**Variable dependiente (a explicar):**
1. **tasa_desempleo**: número de personas desocupadas con respecto a la fuerza total de trabajo del país. Se obtiene mediante la ENE (Encuesta Nacional de empleo).

####**Variables independientes relacionadas con actividad económica:**

2. **imacec_empalmado:** Indicador Mensual de Actividad Económica en Chile es un  índice relativo que refleja el crecimiento o la disminución de la actividad económica en comparación con un período de referencia (índice 2018=100), resumiendo
la actividad de los distintos sectores económicos en un determinado mes, a precios del año anterior encadenados.

3. **imacec_produccion_bienes:** componente específico del IMACEC que se enfoca en medir la actividad económica del sector dedicado a producir bienes (minería, industria y otros). Este indicador está expresado como variación porcentual de IMACEC de producción de bienes respecto a mes anterior.


4. **imacec_comercio:** componente específico del IMACEC que se enfoca en medir la actividad económica del sector comercio. Este indicador está expresado como variación porcentual de IMACEC comercio respecto a mes anterior.

5. **imacec_servicios:** componente específico del IMACEC que se enfoca en medir la actividad económica del sector de servicios. Este indicador está expresado como variación porcentual de IMACEC servicios respecto a mes anterior.

6. **imacec_impuestos:** componente específico del IMACEC que se enfoca en medir la actividad de los impuestos relacionados a productos y servicios (IVA, derechos de importación e impuestos específicos).. Este indicador está expresado como variación porcentual de IMACEC impuestos respecto a mes anterior.

7. **imacec_no_minero:** componente específico del IMACEC que se enfoca en medir la actividad de los sectores no relacionados con minería. Este indicador está expresado como variación porcentual de IMACEC no minero respecto a mes anterior.


####**Variable independiente relacionada con política monetaria:**

8. **TPM:** unidad expresada en pocentaje que refleja la tasa de interés clave establecida por el Banco Central de Chile como parte de su política monetaria. El Banco Central ajusta esta tasa para influir en las condiciones financieras y económicas del país. Por ejemplo, cuando se dice que la TPM es del 3%, significa que la tasa de interés es del 3 por ciento. Este valor determina el costo del endeudamiento y, por lo tanto, tiene un impacto en la economía, el crecimiento y la inflación en Chile.

####**Variables independientes relacionadas con expectativas económicas:**
9. **expectativa_inflacion**: métrica mensual que refleja la variación inflacionaria respecto a periodo anterior. Se obtiene mediante la Encuesta de Expectativas Económicas, encuesta mensual que se realiza a un grupo de académicos, consultores y ejecutivos o asesores de instituciones financieras, la cual entrega información de expectativas de distintas variables macroeconómicas. El dato publicado corresponde a la mediana de las respuestas recibidas.

10. **expectativa_económica:** índice de confianza del consumidor de IPSOS, mide el nivel de optimismo de los consumidores sobre el estado de la economía a través de las actitudes del estado actual y futuro de las economías locales, las finanzas personales, los ahorros y la confianza para realizar inversiones. La métrica es obtenida a través de alrededor de 1.000 encuestas que se efectúan  cada mes en Chile, además de considerar también los dos
meses anteriores, consistiendo en “promedios móviles” ponderados. El rango de este índice va de 0 a 100.

11. **IMCE**: Índice mensual de confianza empresarial. El IMCE se elabora a partir de las respuestas de una encuesta mensual a 600 ejecutivos de empresas de la industria manufacturera, minería, construcción y comercio. La muestra está formada por un panel de empresas, lo que significa que se encuesta a las mismas instituciones en cada levantamiento de datos. Es un índice sintético de difusión que se construye con la suma ponderada de los indicadores de las 4 áreas mencionadas. El indicador queda distribuido en el rango [0, 100], donde el valor 50 marca el punto neutral. Los resultados sobre 50 son optimistas, y bajo 50 son pesimistas.

####**Variables independientes relacionadas con volumen de avisos laborales:**
12. **avisos_laborales:** IALI (Índice de Avisos Laborales de Internet) es un índice de base 100 (en Enero 2015) que refleja el crecimiento o la disminución del volumen de avisos laborales en la web.

### <font color='blue'>5. Análisis de multicolinealidad: </font>

Para obtener un buen modelo es preciso evitar la multicolinealidad entre variables independientes, que se produce cuando hay una fuerte correlación entre ellas. La presencia de multicolinealidad puede incrementar la varianza de los coeficientes de regresión, haciéndolos inestables, y en ese caso no podremos aislar la relación entre cada variable independiente y la variable dependiente.

Para revisar la presencia de multicolinealidad en nuestro dataset ocuparemos la métrica VIF (Variance Inflation Factor). El VIF se calcula para cada variable independiente en el modelo y un valor alto indica que la variable en cuestión está altamente correlacionada con otras variables independientes, lo que hace que sea difícil distinguir su efecto único en el modelo.

Un VIF de 1 indica que no hay multicolinealidad entre la variable y las demás variables independientes. Cuanto mayor sea el VIF, mayor será la multicolinealidad.

Consideraremos que un VIF superior a 10 implica multicolinealidad problemática. Cuando se detecta multicolinealidad.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Colocamos en X todas las variables dependientes:
X = desempleo1[['avisos_laborales', 'imacec_produccion_bienes', 'imacec_comercio', 'imacec_servicios', 'imacec_impuestos', 'imacec_no_minero', 'IMACEC_empalmado', 'expectativa_inflacion', 'IMCE', 'expectativa_economica', 'TPM']]

# Agregar una constante al modelo
X = sm.add_constant(X)

# Calcular VIF para cada variable
vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

Al chequear los resultados del índice VIF para las variables, se tiene que "imacec_servicios" e "imacec_no_minero" están altamente correlacionadas con el resto de las columnas (ambas presentan un VIF sobre 10), por lo que al considerarlas en una regresión sería complicado aislar su efecto indivudual en el desempleo. Por ellos se procederá a eliminarlas del dataset.

```python
desempleo1 = desempleo1.drop(['imacec_servicios', 'imacec_no_minero'], axis=1)
desempleo1.head()
```

## <font color='blue'>**IV. Modelamiento**</font>
### <font color='blue'>1. Modelo Inicial propuesto: Desempleo vs Actividad Económica y TPM</font>

Como se describió en la pregunta de investigación, se generará un modelo de regresión lineal múltiple tipo OLS (mínimo cuadrados ordinarios) que explique la tasa de desempleo en función de la actividad económica y la TPM.

Recordemos que el método de OLS busca encontrar la "línea de regresión" que minimiza la suma de los cuadrados de las diferencias entre los valores observados de la variable dependiente y los valores predichos por el modelo.

#**Modelo 1: DESEMPLEO VS TPM y ACTIVIDAD ECONÓMICA**

```python
# Estimamos un modelo MPL (Regresión por mínimos cuadrados ordinarios) Desempleo vs distintos IMACEC y TPM

X_mpl1 = desempleo1.loc[:,['TPM', "imacec_produccion_bienes",   "imacec_comercio",     "imacec_impuestos", "IMACEC_empalmado",]]
X_mpl1 = sm.add_constant(X_mpl1)
y_mpl1 = desempleo1['tasa_desempleo']

mpl_model1 = sm.OLS(y_mpl1, X_mpl1)
mpl_result1 = mpl_model1.fit()

# Imprimimos el resumen del modelo MPL
print(mpl_result1.summary())
```

Con este modelo obtuvimos un R cuadrado de 0,29, es decir, con este modelo el 29% de la variación de la tasa de desempleo se puede explicar con los 7 regresores escogidos. Este indicador resulta ser más bajo del esperado, teniendo en consideración que la hipótesis inicial planteaba que estas variables deberían tener un mayor efecto en al desocupación.
